```python
import requests
import pandas
```

```python
url = "https://dummyjson.com/carts"
```

```python
carts_data = requests.get(url)
```

```python
carts_data
```

```python
carts_list = carts_data.json()

rows = []
for cart in carts_list["carts"]:
    for product in cart["products"]:
        rows.append({
            "cart_id": cart["id"],
            "userId": cart["userId"],
            "cart_total": cart["total"],
            "cart_discountedTotal": cart["discountedTotal"],
            "cart_totalProducts": cart["totalProducts"],
            "cart_totalQuantity": cart["totalQuantity"],
            **product
        })

subset_df = pandas.DataFrame(rows)
```

```python
subset_df
```

```python
subset_df.to_csv("carts.csv")
```
